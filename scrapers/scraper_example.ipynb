#### Defining file paths

```python
in_filepath = "../raw_data/notas_example.tsv"
out_filepath = "../formatted_data/example.tsv"
```

#### Extract data from raw .tsv

```python
lines = []
with open(in_filepath, "r") as f:
            curr = f.readline() # skips first line
            curr = f.readline()
            while curr != "":
                curr_tok = curr.split("\t")
                curr_fgrade = curr_tok[-1].split()[0]
                try:
                    t1 = float(curr_tok[3].replace(',', '.'))
                except Exception:
                    t1 = 0.0
                try:
                    t2 = float(curr_tok[4].replace(',', '.'))
                except Exception:
                    t2 = 0.0
                if curr_fgrade.isdigit():
                    final_grade = int(curr_tok[-1])
                else:
                    final_grade = -1
                grade = (t1 + t2)*0.5
                lines.append((curr_tok[0], round(grade, 2), final_grade))
                curr = f.readline()
print(lines)
```

#### Write to formatted .tsv file

```python
# insert lines
with open(out_filepath, "w") as f:
    for line in lines:
        name, grade, final_grade = line
        f.write("\t".join([name, str(grade), str(final_grade)]) + '\n')
```
